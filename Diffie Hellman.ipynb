# **Diffie-Hellman Key Exchange**

Αυτό το εργαστήριο θα δημιουργήσει ένα ζεύγος κλειδιών για εσάς, από το οποίο μπορείτε να στείλετε το δημόσιο κλειδί σε έναν συνεργάτη. Μόλις ο συνεργάτης σας σας στείλει το δημόσιο κλειδί του, επικολλήστε το στο πλαίσιο του δημόσιου κλειδιού του. Εάν το κάνετε με επιτυχία, εσείς οι δύο θα πρέπει να έχετε ένα ίδιο κοινό μυστικό. Μπορείτε να χρησιμοποιήσετε το κοινό μυστικό για να κρυπτογραφήσετε μηνύματα μεταξύ τους, πιθανώς χρησιμοποιώντας το εργαλείο AES String Encryption and Decryption.

Το OpenSSL μπορεί να σας βοηθήσει να πραγματοποιήσετε μια ανταλλαγή κλειδιών Diffie-Hellman, αλλά δεν είναι άμεσα συμβατό με αυτό το εργαλείο. Η αρχή όμως είναι η ίδια.

Κατά τη διάρκεια αυτής της διαδικασίας, θα χρειαστεί να δημιουργήσουμε 5 στοιχεία πριν εξαγάγουμε ένα κοινό μυστικό:

* Κοινή βάση (παράμετροι)
* Το ιδιωτικό κλειδί του συνεργάτη 1
* Δημόσιο κλειδί συνεργάτη 1
* Το ιδιωτικό κλειδί του συνεργάτη 2
* Δημόσιο κλειδί του Partner 2

# **Δημιουργήστε μια κοινή βάση**
Χρησιμοποιήστε αυτήν την εντολή για να δημιουργήσετε μια κοινή βάση, dhp.pem. Αυτό το κλειδί παραμέτρων DH μπορεί να κοινοποιηθεί με δημόσιο τρόπο μεταξύ των δύο μερών. Δεν είναι μυστικό. Η βιβλιοθήκη που χρησιμοποιείται από το CryptoTools.net χρησιμοποιεί μια σταθερή τιμή για τη βάση της.

```python
!openssl genpkey -genparam -algorithm DH -out dhp.pem
```

# **Ζεύγος κλειδιών συνεργάτη 1**
Ο Συνεργάτης 1 θα πρέπει να έχει στην κατοχή του το αρχείο dhp.pem και στη συνέχεια να μπορεί να δημιουργήσει τα ιδιωτικά και δημόσια κλειδιά του, dhpriv1.pem και dhpub1.pem, αντίστοιχα.

```python
!openssl genpkey -paramfile dhp.pem -out dhpriv1.pem
!openssl pkey -in dhpriv1.pem -pubout -out dhpub1.pem
```

# **Ζεύγος κλειδιών Partner 2**
Ομοίως, ο Συνεργάτης 2 θα πρέπει να έχει το ίδιο αρχείο dhp.pem και να δημιουργεί τα δικά του ιδιωτικά και δημόσια κλειδιά.

```python
! openssl genpkey -paramfile dhp.pem -out dhpriv2.pem
! openssl pkey -in dhpriv2.pem -pubout -out dhpub2.pem
```

# **Εξάγετε το κοινό μυστικό**
Ο Συνεργάτης 1 θα πρέπει να μοιράζεται το dhpub1.pem με τον Συνεργάτη 2 και ο Συνεργάτης 2 θα πρέπει να μοιράζεται το dhpub2.pem με τον Συνεργάτη 1. Αφού ανταλλάξουν δημόσια κλειδιά, ενδέχεται να αντλήσουν το κοινό μυστικό. Το αντίγραφο του κοινόχρηστου μυστικού του συνεργάτη 1 θα ονομάζεται shared1.bin και του συνεργάτη 2 θα ονομάζεται shared2.bin.

```python
!openssl pkeyutl -derive -inkey dhpriv1.pem -peerkey dhpub2.pem -out shared1.bin
```

Το περιεχόμενο των αρχείων shared1.bin και shared2.bin πρέπει να είναι ακριβώς το ίδιο. Δεδομένου ότι είναι δυαδικά αρχεία και δεν είναι εύκολα αναγνώσιμα από τον άνθρωπο, μπορείτε να επαληθεύσετε ότι είναι ίδια όπως:

```python
!sha256sum shared*.bin
```
