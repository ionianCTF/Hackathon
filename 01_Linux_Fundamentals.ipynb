# **Introduction to Networking**

Networking forms the backbone of modern communication systems, enabling devices to connect and communicate with each other over various types of networks. Two fundamental components in networking are IP addresses and MAC addresses.

---



**IP Address:** An IP address, or Internet Protocol Address, is a unique identifier assigned to devices on a network. It allows devices to communicate with each other by specifying the source and destination of data packets. IP addresses play a crucial role in routing data across networks and connecting devices worldwide.

**MAC Address:** A MAC address, or Media Access Control Address, is a hardware address assigned to network interfaces. It provides a unique identifier for devices at the data link layer of the OSI model. MAC addresses are essential for local network communication and help in distinguishing devices on the same network.

---
The ip addr command in Linux (or ifconfig and ipconfig on Windows OS) provides detailed information about network interfaces, including essential details such as interface names (e.g., eth0, wlan0), MAC addresses, IPv4 and IPv6 addresses, subnet masks, and more. This command is crucial for network configuration and troubleshooting. Each indicator on the interface names is a different network interface. The interface names are different on Windows and OSX operating systems. Therefore it is important to understand what each network interaface is (e.g. wlan0, eth0, virtual-interfaces etc.).

Obtaining the IP address is crucial in cybersecurity as the first step due to the following reasons. IP addresses help identify devices connected to a network. By obtaining the IP address of a device, cybersecurity professionals can pinpoint the source of potential threats or suspicious activities.

Here is an explanation of the key terms and details from the provided output of the ip addr command:

**lo (Loopback Interface)**
*   lo (Loopback Interface), LOOPBACK: Indicates that this interface is the loopback interface, allowing a system to communicate with itself.
*   LOWER_UP: Indicates that the link is physically up.
* MTU (Maximum Transmission Unit): Specifies the largest packet size that can be transmitted.
* inet 127.0.0.1/8: This line shows the IPv4 address 127.0.0.1 assigned to the loopback interface with a subnet mask of /8.
* valid_lft forever preferred_lft forever: Indicates that the address is valid indefinitely.

**eth0**
* BROADCAST: Indicates that the interface supports broadcast packets.
* MULTICAST: Shows that the interface supports multicast packets.
* UP: Indicates that the interface is active and operational.
* LOWER_UP: Indicates that the link is physically up.
* MTU (Maximum Transmission Unit): Specifies the largest packet size that can be transmitted.
* inet 172.28.0.12/16: This line displays the IPv4 address 172.28.0.12 assigned to the Ethernet interface with a subnet mask of /16.
* brd 172.28.255.255: Specifies the broadcast address for the subnet.
* Scope: global: Indicates that the address is reachable from anywhere on the network.
